#### By using Pandas' describe function, we are able to get a sense of understanding of the numerical data in the dataframe. We can see the average value of rent across all properties has increased by 133 from January 2011 to December 2016. We can also notice that there is an even lower rent value in that same time frame. It is important to note that there are 'NaN' values for some states. This is either as a result of Zillow not having properties in the state or possibly because the data for those areas were not being tracked.

```python
df['State'].unique()
```

```python
px.histogram(df, x="State").update_xaxes(categoryorder="total descending")
```

##### By plotting this histogram, we are able to see a visual representation which helps us better understand the distribution of our data.

From the histogram we are able to gather that Pennsylvania has the most Zillow houses with data recorded for 1260 different homes, followed by New York with 775 and California with 772

```python
df.groupby('State')['January 2012'].mean()
```
