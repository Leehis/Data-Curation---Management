#### We're going to drop the November 2010, December 2010, and the January 2017 columns so that our data starts and ends at an even time period of 6 years.

```python
new_df = df
new_df.drop(df.columns[[6,7, 80]], inplace=True, axis=1)
new_df
```

```python
new_df['2011'] = df[['January 2011',
                     'February 2011', 'March 2011', 'April 2011', 'May 2011', 'June 2011', 'July 2011', 'August 2011', 'September 2011', 'October 2011', 'November 2011', 'December 2011']].mean(axis=1)

new_df['2012'] = df[['January 2012',
                     'February 2012', 'March 2012', 'April 2012', 'May 2012', 'June 2012', 'July 2012', 'August 2012', 'September 2012', 'October 2012', 'November 2012', 'December 2012']].mean(axis=1)

new_df['2013'] = df[['January 2013',
                     'February 2013', 'March 2013', 'April 2013', 'May 2013', 'June 2013', 'July 2013', 'August 2013', 'September 2013', 'October 2013', 'November 2013', 'December 2013']].mean(axis=1)

new_df['2014'] = df[['January 2014',
                     'February 2014', 'March 2014', 'April 2014', 'May 2014', 'June 2014', 'July 2014', 'August 2014', 'September 2014', 'October 2014', 'November 2014', 'December 2014']].mean(axis=1)

new_df['2015'] = df[['January 2015',
                     'February 2015', 'March 2015', 'April 2015', 'May 2015', 'June 2015', 'July 2015', 'August 2015', 'September 2015', 'October 2015', 'November 2015', 'December 2015']].mean(axis=1)

new_df['2016'] = df[['January 2016',
                     'February 2016', 'March 2016', 'April 2016', 'May 2016', 'June 2016', 'July 2016', 'August 2016', 'September 2016', 'October 2016', 'November 2016', 'December 2016']].mean(axis=1)

new_df
```

#### We have averaged the values of the months in each year to find the average value of rent on a yearly basis

```python
new_df_year = new_df
new_df_year = new_df_year.drop(new_df_year.loc[:, 'January 2011':'December 2016'].columns, axis=1)

new_df_year
```

```python
new_df_index_updated = new_df_year.set_index('City Code')
new_df_index_updated
```

```python
new_df_index_updated['2012 Change'] = ((new_df_index_updated['2012'] - new_df_index_updated['2011']) / new_df_index_updated['2011']) * 100
new_df_index_updated['2013 Change'] = ((new_df_index_updated['2013'] - new_df_index_updated['2012']) / new_df_index_updated['2012']) * 100
new_df_index_updated['2014 Change'] = (
    (new_df_index_updated['2014'] - new_df_index_updated['2013']) / new_df_index_updated['2013']) * 100
new_df_index_updated['2015 Change'] = (
    (new_df_index_updated['2015'] - new_df_index_updated['2014']) / new_df_index_updated['2014']) * 100
new_df_index_updated['2016 Change'] = (
    (new_df_index_updated['2016'] - new_df_index_updated['2015']) / new_df_index_updated['2015']) * 100




new_df_index_updated
```

```python
new_york_df = new_df_index_updated.loc[new_df_index_updated['State'] == 'NY']


new_york_df
```

#### We're going to create a graphic to identify trends about the monthly rent values across the years for counties and metros in the State of New York.
##### In order to do this we're going to have to sum up the values across all the months in a year into a new column to that corresponding year

```python
ny_by_county = new_york_df.groupby('County').mean()
ny_by_county_years = ny_by_county.drop(columns=['Population Rank', '2012 Change', '2013 Change', '2014 Change', '2015 Change', '2016 Change'])
ny_by_county_years
```

```python
#ny_by_county.stack()
ny_by_county_years = ny_by_county_years.T
ny_by_county_years
```

```python
fig = px.line(ny_by_county)
fig.show()
```

#### Identifying Trends
From the line graph above, there are a lot of trends we can see. 
Between 2012 and 2011
Some counties saw a decrease in the average price of rent such as Niagara with a -25% drop in value, Broome with -19%, Greene with -15% and Rensselaer with -12% taking the biggest hits.
Madison county saw the greatest increase with a surge of 5.05%
